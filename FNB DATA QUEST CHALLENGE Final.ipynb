```python
df['int_date'] = pd.to_datetime(df['int_date'], format='%d%b%Y')
```

```python
fd = df[df['item'] != 'NONE']
```

```python
fd.head(30)
```

```python
interaction_score = {'DISPLAY': 1, 'CLICK': 2, 'CHECKOUT': 3}
fd['interaction_score'] = fd['interaction'].map(interaction_score)
```

```python
user_item_score = fd.groupby(['idcol', 'item']).agg({'interaction_score': 'sum'}).reset_index()
```

```python
scaler = MinMaxScaler()
user_item_score['interaction_score'] = scaler.fit_transform(user_item_score[['interaction_score']])
```

```python
train_data, test_data = train_test_split(user_item_score, test_size=0.2, random_state=42)
```

```python
interaction_matrix = train_data.pivot(index='idcol', columns='item', values='interaction_score').fillna(0)
```

```python
interaction_sparse_matrix = csr_matrix(interaction_matrix.values)
```

```python
print(type(interaction_sparse_matrix))
```

```python
U, sigma, Vt = svds(interaction_sparse_matrix, k=50)
sigma = np.diag(sigma)
```

```python
predicted_scores = np.dot(np.dot(U, sigma), Vt)
predicted_scores_df = pd.DataFrame(predicted_scores, columns=interaction_matrix.columns, index=interaction_matrix.index)
```

```python
def recommend_items(idcol, num_recommendations=5):
    if idcol in predicted_scores_df.index:
        user_predictions = predicted_scores_df.loc[idcol].sort_values(ascending=False)
        top_items = user_predictions.head(num_recommendations).index.tolist()
    else:
        # Handle cold start by recommending popular items or a default set
        top_items = fd['item'].value_counts().head(num_recommendations).index.tolist()
    return top_items
```

```python
def get_rmse(predicted, actual, sparse_matrix):
    predicted_flatten = predicted[sparse_matrix.nonzero()].flatten()
    actual_flatten = actual[sparse_matrix.nonzero()].flatten()
    return np.sqrt(mean_squared_error(predicted_flatten, actual_flatten))
```

```python
# Compute RMSE on the training set
actual_scores = interaction_matrix.values
print('RMSE:', get_rmse(predicted_scores, actual_scores, interaction_sparse_matrix))
```
